```python
import os 
import pandas as pd
import json
import plotly.express as px
import plotly.graph_objects as go
```

```python
def open_all(foldername : str) -> pd.DataFrame: 
    
    results = {}
    
    for current_dir in os.listdir(f'{foldername}'):
        results["train_loss"] = pd.read_csv(
            f'{foldername}/{current_dir}/train_loss.csv')['loss_value'].tolist()
        
        results["valid_loss"] = pd.read_csv(
            f'{foldername}/{current_dir}/validation_loss.csv')['loss_value'].tolist()
        results["confusion_matrix"] = pd.read_csv(
            f'{foldername}/{current_dir}/confusion_matrix.csv'
        )
    return results
```
